```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

```python
# reading in saved pickle file
df = pd.read_pickle('testing.pkl')
```

```python
df.head()
```

```python
df.shape
```

```python
# Subsetting returns on day T
X = df.iloc[:, 27:146]

X.head()
X.shape
```

```python
# Doing a backfill so I avoid dropping any returns column
X.fillna(method='backfill', axis=1, inplace=True)
```

```python
# Calculating the compound sum of returns on day T
X['Ret_comp_sum'] = X.compound(axis=1)
```

```python
# Calculating the standard deviation of returns on day T
X['Ret_std'] = X.std(axis=1)
```

```python
X.head()
```

```python
# Calculating the difference between T minus one and T minus two
X['diff']  = df.Ret_MinusOne - df.Ret_MinusTwo
```

```python
X.head()
```

```python
# Subsetting out masked features
m_feat = df.iloc[:, :25]
```

```python
# Dropping features with more than 50% of data missing
m_feat = m_feat.drop(['Feature_1', 'Feature_10', 'Feature_2', 'Feature_20', 'Feature_4'], axis=1)
```

```python
m_feat.head()
```

```python
# Imputting empty rows in feature set
imp = Imputer(axis=0)
imp.fit(m_feat)
imp_feat = imp.transform(m_feat)
```

```python
# Converting features back to dataframe
m_feat_x = pd.DataFrame(data=imp_feat, columns=m_feat.columns.values)
```

```python
m_feat_x.head()
```

```python
m_feat_x.shape
```

```python
# Combining masked features and returns features together 
df_new = pd.concat([m_feat_x, X], axis=1)
```

```python
df_new.shape
```

```python
df_new.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.linear_model import Ridge, Lasso, ElasticNet, LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler, Imputer
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.decomposition import PCA
from sklearn.metrics import mean_absolute_error
```

```python
# creating a subset of my X train data 
features = X.iloc[:, 119:]
features.head()
```

```python
# defining target
y = df.Ret_PlusOne
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(df_new, y, test_size = 0.3, random_state=42)
```

```python
# defining number of folds to use for cross validation 
cv = KFold(n_splits=10, shuffle=True, random_state=20)
```

```python
# Scaling features and performing a principal component analysis on features
ss = StandardScaler()
ss.fit(X_train)
Xs = ss.transform(X_train)

pca = PCA()
pca.fit(Xs)
Xpca = pca.transform(Xs)
```

```python
# Training on a linear regression model
linreg = LinearRegression()
linreg.fit(Xpca, y_train)

# Not as good as the linear regression score in part 1, slightly better than part 2
cross_val_score(linreg, Xpca, y_train, cv=cv).mean()
```

```python
mean_absolute_error(y_train, linreg.predict(Xpca))
```

```python
# Doing a grid search over an elastic error reduction model
elastic_params = {
    'alpha' : np.logspace(-5, 5, 20),
    'l1_ratio' : [0.1, 0.2, 0.5, 0.6, 0.8, 0.9]
}

elastic_gridsearch = GridSearchCV(ElasticNet(), elastic_params, verbose=1, cv=10)
```

```python
elastic_gridsearch.fit(Xpca, y_train)
best_elastic = elastic_gridsearch.best_estimator_
```

```python
# Not as good as the ridge scores in part 1, but slightly better than part 2 
elastic_gridsearch.best_score_
```
