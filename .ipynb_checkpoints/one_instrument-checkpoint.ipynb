```python
import pandas_datareader.data as web
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import timedelta
from __future__ import division
plt.style.use('ggplot')
%matplotlib inline
```

```python
# setting  start and end time for which I want to collect my data
start = datetime.datetime(2010, 9, 6)
end = datetime.datetime(2017, 9, 1)
```

```python
# obtaining amazon stock prices within this period
amzn = web.DataReader('AMZN', 'yahoo', start, end)
```

```python
amzn.head()
```

```python
amzn.shape
```

```python
# Calculating returns
amzn['returns'] = amzn['Adj Close'].pct_change()
```

```python
amzn.head()
```

```python
# Doing a two day back shift to simulate the structure of the data I was provided for the challenge
for i in range(1,3):
    varname = "shift-"+str(i)
    amzn[varname] = amzn["returns"].shift(i)
```

```python
amzn.head()
```

```python
# Doing a two day forward shift to simulate the structure of the data I was provided for the challenge
for i in range(-2,0):
    varname = "shift+"+str(i)
    amzn[varname] = amzn["returns"].shift(i)
```

```python
amzn.head()
```

```python
amzn.shape
```

```python
amzn.isnull().sum()
```

```python
df = amzn.copy()
```

```python
# dropping null values
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# plotting the adjusted close values for the time period
fig, ax = plt.subplots(figsize=(16,8))

ax.set_title('AMZN Daily Closing Price')
df['returns'].plot();
```

```python
# making my features and target
y = df['shift+-1']
X = df[['shift-1', 'shift-2', 'returns']]
```

```python
train_X = X['2010-09-09':'2016-12-30']
train_y = y['2010-09-09':'2016-12-30']

test_X = X['2017-01-01':'2017-08-30']
test_y = y['2017-01-01':'2017-08-30']
```

```python
from sklearn.linear_model import Lasso, LinearRegression
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import mean_squared_error

cv = KFold(n_splits=10, shuffle=True, random_state=20)

lr = Lasso()
lr.fit(train_X, train_y)
# print lr.score(train_X, train_y)
print mean_squared_error(train_y, lr.predict(train_X))
cross_val_score(lr, train_X, train_y, cv=cv).mean()
```

```python
lr.score(test_X, test_y)
```
