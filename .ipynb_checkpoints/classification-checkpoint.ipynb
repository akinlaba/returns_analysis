### Showing results of only my best performing data structure

```python
# reading in saved pickle file
df = pd.read_pickle('testing.pkl')
```

```python
df.head()
```

```python
# Subsetting returns on day T
X = df.iloc[:, 27:146]

X.head()
X.shape 
```

```python
# Doing a backfill so I avoid dropping any returns column
X.fillna(method='backfill', axis=1, inplace=True)
```

```python
def compound(percent_returns):
    '''Function that takes in a row and returns a list of the cumulative \
    sum of each record in the row from ret 2 to ret 120'''
    compound_sum = []
    total = 0
    for i in percent_returns:
        total += i
        compound_sum.append(total)
        
    return compound_sum
```

```python
# Calling the compound function on returns on day T
comp_features = X.apply(compound,axis=1)
```

```python
comp_features.columns
```

```python
# Renaming column headers
comp_features.columns = [feats.replace("Ret", "Comp") for feats in comp_features.columns]
```

```python
comp_features.head()
```

```python
# Calculating the standard deviation of returns on day T
comp_features['Ret_std'] = X.std(axis=1)
```

```python
# Calculating the difference between T minus one and T minus two
comp_features['diff']  = df.Ret_MinusOne - df.Ret_MinusTwo
```

```python
# defining target
y = pd.DataFrame(df.Ret_PlusOne)
```

```python
# converting target into labels
y['target'] = y.Ret_PlusOne.map(lambda x: 1 if x > 0 else 0)
```

```python
y.head()
```

```python
# Baseline
y.target.mean()
```

```python
comp_features.shape
```

```python
# Train test split
X_train, X_test, y_train, y_test = train_test_split(comp_features, y.target, test_size = 0.3, random_state=42)
```

```python
# defining number of folds to use for cross validation 
cv = KFold(n_splits=10, shuffle=True, random_state=20)
```

```python
# Scaling features and performing a principal component analysis on features
ss = StandardScaler()
ss.fit(X_train)
Xs = ss.transform(X_train)

pca = PCA()
pca.fit(Xs)
Xpca = pca.transform(Xs)
```

```python
# Plotting principal components explained to determine number of features to use
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(range(1,122), pca.explained_variance_ratio_)
ax.scatter(range(1, 122), pca.explained_variance_ratio_)
ax.set_title('explained variance of components')
ax.set_xlabel('principal component')
ax.set_ylabel('explained variance')
```

```python
# Using top 12 principal components
pca_df = pd.DataFrame(Xpca).iloc[:,:12]
```

```python
pca_df.head()
```

```python
# Training using only standardized features on a KNN classifier
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(Xs, y_train)

# Not doing that well over the baseline
cross_val_score(knn, Xs, y_train, cv=cv).mean()
```

```python
# Training using pca features on a KNN classifier
knn2 = KNeighborsClassifier(n_neighbors=10)
knn2.fit(pca_df, y_train)

# Doing worse than using only standardized features
cross_val_score(knn2, pca_df, y_train, cv=cv).mean()
```

```python
# Defining stochastic gradient classifier parameters
sgd_cls_params = {
    'loss':['log'],
    'penalty':['l1','l2'],
    'alpha':np.logspace(-5,2,50)
}

sgd_cls = SGDClassifier()
sgd_cls_gs = GridSearchCV(sgd_cls, sgd_cls_params, cv=cv, verbose=1)
```

```python
# Training on SGD model
sgd_cls_gs.fit(Xs, y_train)
```

```python
# Not doing that better than the baseline
print sgd_cls_gs.best_params_
print sgd_cls_gs.best_score_
```

```python
# training on a decision tree classifier
clf = DecisionTreeClassifier(max_depth =None)
clf.fit(Xs, y_train)

# scoring and not performing better than the baseline
cross_val_score(clf, Xs, y_train, cv=cv).mean()
```

```python
# training on a bagging tree classifier
bdt = BaggingClassifier(DecisionTreeClassifier())
bdt.fit(Xs, y_train)

# scoring and not performing better than the baseline
sbdt = cross_val_score(bdt, Xs, y_train, cv=cv)
print("{} Score:\t{:0.3}  {:0.3}".format("Bagging", sbdt.mean().round(3), sbdt.std().round(3)))
```

```python
# training on a random forest classifier
rf = RandomForestClassifier(n_estimators=10, n_jobs=-1)
rf.fit(Xs, y_train)

# scoring and not performing better than the baseline
srf = cross_val_score(rf, Xs, y_train, cv=cv)
print("{} Score:\t{:0.3}  {:0.3}".format("Random", srf.mean().round(3), srf.std().round(3)))
```

```python
# training on a extra tree classifier
etc = ExtraTreesClassifier(n_jobs=-1)
etc.fit(Xs, y_train)

# scoring and not performing better than the baseline
setc = cross_val_score(etc, Xs, y_train, cv=cv)
print("{} Score:\t{:0.3}  {:0.3}".format("Decision Tree", setc.mean().round(3), setc.std().round(3)))
```
